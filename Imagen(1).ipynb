Una vez redimensionada cargarla en el colab para predecir con el modelo en cuestion

```python
from skimage.transform import resize
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
images=[]
# AQUI ESPECIFICAMOS UNAS IMAGENES
filenames = ['5(1).png']

for filepath in filenames:
    image = plt.imread(filepath,0)
    image_resized = resize(image, (28, 28),anti_aliasing=True,clip=False,preserve_range=True)
    images.append(image_resized)
```

```python
X = np.array(images, dtype=np.uint8) #convierto de lista a numpy
test_X = X.astype('float32')#convierto a float32
test_X = test_X / 255. #normalizo a 0-1
```

```python
import keras
```

```python
model = keras.models.load_model('modelmnist.h5')
```

```python
predicted_classes = model.predict(test_X)
```

```python
predicted_classes
```

```python
asd= predicted_classes.argmax()
```
