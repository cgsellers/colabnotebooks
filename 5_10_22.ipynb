```python
import pandas as pd
```

```python
datos1= '/content/datos1.csv'
```

```python
df1= pd.read_csv(datos1)
df1
```

```python
df1['SEXO']=pd.Categorical(df1['SEXO'])
df1['SEXO']
```

```python
df1['PAIS']=pd.Categorical(df1['PAIS'])
df1['TRABAJO']=pd.Categorical(df1['TRABAJO'])
```

```python
df1.info()
```

```python
sexo_dummy = pd.get_dummies(df1['SEXO'],prefix='SEXO')
```

```python
df1=pd.concat([df1,sexo_dummy],axis=1)
```

```python
df1
```

```python
df1=df1.drop(['SEXO'],axis=1) #quitar la columna sexo
```

```python
df1
```

```python
df1['PAIS']=df1['PAIS'].cat.codes
```

```python
df1.info()
```

```python
#para usar label encoder tenemos que utilizar la libreria numpy
import numpy as np
from sklearn.preprocessing import LabelEncoder #Creating initial dataframe
```

```python
labelencoder = LabelEncoder()
df1['TRABAJO']= labelencoder.fit_transform(df1['TRABAJO']) #Usamos el metodo fit transform del labelencoder para pasar de categorical a int64
```
