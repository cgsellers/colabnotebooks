```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('fast')

from keras.models import Sequential
from keras.layers import Dense,Activation,Flatten
from sklearn.preprocessing import MinMaxScaler
fich='Temporal.csv'
aa18=pd.read_csv(fich)
```

```python
X = aa18.drop('Valor', axis=1)
Y = aa18['Valor']
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
# Normalizamos variables
scaler = StandardScaler()
XY = scaler.fit_transform(X)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(XY, Y, test_size=0.2, random_state=0)
```

#NN en keras

```python
from keras.models import Sequential
from keras.layers import Dense

# Crear modelo NN
model = Sequential()
model.add(Dense(units=1, activation='relu', input_dim=1))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model.fit(X_train, Y_train, epochs=10, batch_size=32)
```

```python
aa=model.predict([[1.4]])
```

```python
aa
```

#RNN en keras

```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense

# Crear modelo RNN
model1 = Sequential()
model1.add(SimpleRNN(units=1, activation='relu', input_shape=(10, 1)))
model1.add(Dense(units=32, activation='relu'))
model1.add(Dense(units=1, activation='sigmoid'))

# Compilar el modelo
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model1.fit(X_train, Y_train, epochs=10, batch_size=32)
```

```python
aa1=model1.predict([[1.4]])
```

```python
aa1
```
