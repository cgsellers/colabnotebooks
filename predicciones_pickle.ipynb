```python
import pickle
import numpy as np
```

#Regresión logística

```python
reg_log="pickle_model_reg_log.pkl"
with open(reg_log, 'rb') as file:
    pickle_model = pickle.load(file)
```

```python
X_axis = np.linspace(1,2, 14).reshape(1, 14)
prediccion=pickle_model.predict(X_axis)
prediccion[0]
```
