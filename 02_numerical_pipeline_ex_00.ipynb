Let's load the adult census dataset with only numerical variables

```python
import pandas as pd
# Colab: https://github.com/INRIA/scikit-learn-mooc/raw/main/datasets/adult-census-numeric.csv
adult_census = pd.read_csv("https://github.com/INRIA/scikit-learn-mooc/raw/main/datasets/adult-census-numeric.csv")
data = adult_census.drop(columns="class")
target = adult_census["class"]
```

In the previous notebook we used `model = KNeighborsClassifier()`. All
scikit-learn models can be created without arguments. This is convenient
because it means that you don't need to understand the full details of a
model before starting to use it.

One of the `KNeighborsClassifier` parameters is `n_neighbors`. It controls
the number of neighbors we are going to use to make a prediction for a new
data point.

‚ùì Create a `KNeighborsClassifier` model with `n_neighbors=50`

```python
# Write your code here.
modelo=KNeighborsClassifier(n_neighbors=50)
```

‚ùì Fit this model on the data and target loaded above

```python
# Write your code here.
modelo.fit(data,target)
```

‚ùì Use your model to make predictions on the first 10 data points inside the
data. Do they match the actual target values?

```python
# Write your code here.
predicciones=modelo.predict(data.iloc[:10])
for prediccion,valor in zip(list(predicciones),target.iloc[:10]):
  print("Predicci√≥n: {} Valor: {}".format(prediccion,valor))
```

‚ùì Compute the accuracy on the training data.

```python
aciertos=0
for prediccion,valor in zip(list(predicciones),target.iloc[:10]):
  if(prediccion==valor):
    aciertos+=1

accuracy=aciertos/10*100
accuracy

```

‚ùì Now load the test data from `"../datasets/adult-census-numeric-test.csv"` ([Github Link](https://github.com/INRIA/scikit-learn-mooc/raw/main/datasets/adult-census-numeric-test.csv)) and
compute the accuracy on the test data.

```python
# Write your code here.
test=pd.read_csv("https://raw.githubusercontent.com/INRIA/scikit-learn-mooc/main/datasets/adult-census-numeric-test.csv")
test_target=test["class"]
test_data=test.drop(columns="class")
```

```python
predicciones_test=modelo.predict(test_data)
```

```python
aciertos_test=0
for prediccion,valor in zip(list(predicciones_test),test_target):
  if(prediccion==valor):
    aciertos_test+=1

test_accuracy=round(aciertos_test/len(test_target)*100,2)
test_accuracy
```
