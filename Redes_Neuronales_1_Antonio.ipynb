Las Redes Neuronles tratan de emular al cerebro humano, de forma que son capaces de reproducir algunas tareas que desarrolla.

#2.1) Redes Neuronales

Una Red Neuronal es un conjunto de elementos de procesamiento de la información altamente interconectados, que son capaces de aprender con la información que se les alimenta. 

Estas Redes Neuronales se pueden aplicar a reconocimiento de imágenes, reconocimiento de voz, analisis y filtrado de señales, clasificación discriminación, analisis financiero, predicción dinamica. 


La unidad básica de una red neuronal es una neurona.

Una neurona artificial es un modelo matemático utilizado en el aprendizaje automático y las redes neuronales artificiales. Se basa en la estructura de una neurona biológica, que es la unidad básica del cerebro humano.

Una neurona artificial consta de una entrada o varias, una función de activación, pesos, un sesgo y una o varias salidas. 

La entrada es una señal o conjunto de señales que se alimenta a la neurona.

Los pesos y el sesgo se utilizan para ajustar la señal de entrada y producir una salida.

La función de activación es una función matemática que se aplica a la suma ponderada de las entradas, los pesos y el sesgo. Su propósito es introducir no linealidad en el modelo y permitir que las neuronas artificiales sean capaces de aprender patrones complejos en los datos de entrada.

Una neurona artificial puede tener múltiples entradas y salidas, y se utiliza en conjunción con otras neuronas para construir redes neuronales artificiales. Estas redes pueden utilizarse para tareas como la clasificación, la predicción y el reconocimiento de patrones en los datos.

    Pesos

Los `pesos` en una neurona artificial son valores numéricos que se utilizan para ajustar la señal de entrada y producir una salida. Cada entrada a la neurona tiene un peso asociado, y estos pesos se ajustan durante el proceso de entrenamiento de la red neuronal.

Los pesos iniciales se establecen al azar y luego se ajustan durante el entrenamiento para minimizar la función de`pérdida del modelo. 

El ajuste de los pesos es un proceso iterativo, y el número de iteraciones requeridas para entrenar la red neuronal depende del tamaño de los datos de entrenamiento y de la complejidad del modelo. Una vez entrenada, la red neuronal puede utilizarse para realizar predicciones precisas sobre datos nuevos.

    Sesgo

El `sesgo` (también conocido como `"bias"`en inglés) en una neurona artificial es un valor numérico que se utiliza para ajustar la salida de la neurona. Es un parámetro adicional en la neurona que permite a la función de activación trasladar el rango de salida de la neurona hacia arriba o hacia abajo.

El sesgo se representa como un solo número en la neurona y es un parámetro aprendido durante el entrenamiento de la red neuronal. Su función es permitir que la neurona sea capaz de aprender y producir una salida incluso cuando todas las entradas tienen un valor de cero.

En la práctica, el sesgo es una forma de introducir una "preferencia" en la neurona por ciertos valores de entrada. Por ejemplo, en una red neuronal que se entrena para predecir el precio de una casa en función de su tamaño y edad, el sesgo podría utilizarse para introducir una preferencia por las casas más grandes o más nuevas.

El sesgo es un parámetro crítico en el proceso de entrenamiento de la red neuronal y es importante tener en cuenta su valor óptimo para obtener un mejor rendimiento de la red.

    Función de activación

Las funciones de activación se encuentran en cada neurona de una red neuronal y la utilidad más importante que tienen es indicar cuando una neurona se activa o se apaga, dependiendo de la función de activación que se use, esta tendra ciertos limites, recordemos que primero se calcula la función Z = WX + b y el resultado de esta función se le pasa a la función de activación, esta ultima busca si los datos tienen los patrones que busca la neurona o no los tiene.

Esta función aplica una transformación no lineal a la suma ponderada de las entradas de la neurona, lo que permite que la red neuronal pueda aprender y modelar relaciones no lineales entre las variables de entrada y salida.

    Batch size

Es el numero de datos que tiene cada iteración de un ciclo (epoch), esto es util porqué la red neuronal actualiza los parametros W y b más veces, también cuando se tienen grandes cantidades de datos se necesitan computadoras con más memoria y la red neuronal tarda más en ejecutar cada ciclo, si dividimos los ciclos en iteraciones con un numero de datos más pequeño ya no es necesario cargar todos los datos en la memoria al mismo tiempo y la red neuronal se entrena más rapido.

    Epoch

Este es el numero de veces que se ejecutaran los algoritmos de forwardpropagation y backpropagation. En cada ciclo (epoch) todos los datos de entrenamiento pasan por la red neuronal para que esta aprenda sobre ellos, si existen 10 ciclos y 1000 datos, cada ciclo los 1000 datos pasaran por la red neuronal. Si se especifica el parametro batch size cada ciclo (epoch) tendra más ejecuciones internas, estas ejecuciones se llaman iteraciones, si tenemos un batch size de 100, se tendran 10 iteraciones para completar un ciclo, en cada iteración se ejecutan los algoritmos de forwardpropagation y backpropagation, de esta manera la red neuronal actualiza más veces los parametros W (peso) y b(sesgo).

    Función de pérdida


La `función de pérdida loss function` es una medida de la diferencia entre la salida real y la salida predicha de la red neuronal.

Durante el entrenamiento, se utiliza un algoritmo de optimización para ajustar los pesos de la neurona de manera que se minimice la función de pérdida. El algoritmo más comúnmente utilizado es el descenso de gradiente, que ajusta los pesos en pequeños incrementos basados en la pendiente de la función de pérdida en relación con los pesos actuales.

La función de perdida, también conocida como función de costo, es la función que nos dice que tan buena es la red neuronal, un resultado alto indica que la red neuronal tiene un desempeño pobre y un resultado bajo indica que la red neuronal esta haciendo un buen trabajo. Esta es la función que optimizamos o minimizamos cuando realizamos el backpropagation.

Existen varias funciones matematicas que pueden usarse, la elección de una depende del problema que se esta resolviendo.

Algunas de estas funciones son:

    Cross Entropy Esta función se usa para problemas de clasificación

    Mean Squared Error Esta función se usa para problemas de regresión.

    Learning rate

Este hiperparametro indica que tan largo sera el camino que tome el algoritmo de optimización. Si el valor es muy pequeño la actualización se puede quedar atrapada en un minimo local y los valores de W y b no cambiarian correctamente, igualmente la red neuronal tardara mucho más tiempo en optimizar. Por otro lado si los valores son muy altos la actualización puede pasarse del punto perfecto que sería el minimo global y nunca encontrarlo y aunque el aprendizaje sea más rapido nunca llegara al minimo global.


    Optimizer


Existen diferentes tipos de algoritmos de optimización el más famoso se llama gradient descent sgd y el más usado se llama Adam. 

El algoritmo de optimización SGD (Stochastic Gradient Descent) es uno de los algoritmos más comúnmente utilizados para entrenar redes neuronales.

Su funcionamiento se basa en cuatro pasos:

1.- Paso de forward: Se calcula el resultado de la red neuronal para una entrada dada.

2.- Cálculo del error: Se calcula el error de la red neuronal en función de la salida esperada y la salida real obtenida.

3.- Paso de backward: Se calcula la derivada parcial del error con respecto a cada peso de la red neuronal.

4.- Actualización de los pesos: Se actualizan los pesos de la red neuronal con la derivada parcial menor.

Donde la tasa de aprendizaje es un hiperparámetro que controla la magnitud de la actualización de los pesos.

Se repiten los pasos 1-4 para cada ejemplo de entrenamiento en el conjunto de entrenamiento.

Se repiten los pasos 1-4 durante múltiples épocas hasta que la red neuronal tenga un resultado satisfactorio.

El algoritmo SGD es conocido como "estocástico" porque se utiliza una muestra aleatoria (o "batch") de ejemplos de entrenamiento para actualizar los pesos en cada iteración. Esto puede ayudar a evitar el sobreajuste y a mejorar la generalización de la red neuronal.

Es importante tener en cuenta que existen variantes del algoritmo SGD, como SGD con momento, que pueden ayudar a mejorar el rendimiento de la red neuronal en ciertos casos. Además, la elección de la tasa de aprendizaje y otros hiperparámetros puede ser crucial para el éxito del entrenamiento de la red neuronal.

`Layer o Dense`

Existen tres tipos de capas (layers) en una red neuronal, todas contienen una o más neuronas:

Capa de entrada (Input Layer): Esta capa contiene neuronas que representan los datos que la red neuronal usara para entrenar. El numero de neuronas de esta capa depende del numero de caracteristicas que tengan los datos.

Capa oculta (Hiden Layer): Una red neuronal puede tener varias capas de este tipo, cada una de estas capas contiene neuronas, en una red neuronal tradicional cada una de las neuronas de una capa estan conectadas con todas las neuronas de la siguiente capa.

Capa de salida (Output Layer): Esta capa es la que se encarga de entregar los resultados, si estamos resolviendo un problema de clasificación esta capa tendra un numero de neuronas igual al numero de clases que existan en los datos. El resultado es una lista de probabilidades para cada clase.
