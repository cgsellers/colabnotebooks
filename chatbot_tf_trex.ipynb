```python
def load_ancora_corpus(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    return [line.strip() for line in lines if not line.startswith("#") and line.strip()]

train_file = "UD_Spanish-AnCora/es_ancora-ud-train.conllu"
valid_file = "UD_Spanish-AnCora/es_ancora-ud-dev.conllu"
test_file = "UD_Spanish-AnCora/es_ancora-ud-test.conllu"

train_dataset = load_ancora_corpus(train_file)
valid_dataset = load_ancora_corpus(valid_file)
test_dataset = load_ancora_corpus(test_file)

# Combina los conjuntos de datos
complete_dataset = train_dataset + valid_dataset + test_dataset
```

```python
complete_dataset
```

```python
nlp = spacy.load("es_core_news_sm")
```

```python
def preprocess_text(text):
    # Procesa el texto con el modelo en español de spaCy
    doc = nlp(text)

    # Convierte el texto a minúsculas y elimina signos de puntuación y espacios adicionales
    tokens = [token.text.lower() for token in doc if not token.is_punct and not token.is_space]

    return tokens
```

```python
preprocessed_dataset = [preprocess_text(sentence) for sentence in complete_dataset]
```
