---
---


```python
#Importamos Librerías
import tensorflow as tf
import numpy as np
```

```python
# Datos de multiplicar del 5
entrada=np.array([[5,0],[5,1],[5,2],[5,3],[5,4],[5,5],[5,6],[5,7],[5,8],[5,9],[5,10]], "int64")

#salida = tf.constant([[0], [5], [10], [15],[20], [25], [30], [35],[40], [45], [50]], dtype=tf.int64)
salida = np.array([0,5,10,15,20,25,30,35,40,45,50], "int64")
```

```python
# Definición del modelo
model = tf.keras.Sequential()
#modelo con 1 neurona y dos entradas
model.add(tf.keras.layers.Dense(1,  input_shape=(2,)))
```

```python
# Compilación del modelo
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
# Entrenamiento del modelo
historial=model.fit(entrada, salida, epochs=1000, verbose=1)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
# Utiliza el modelo para hacer predicciones
x_test = np.array([[5,11],[5,12],[5,13],[5,14]])
y_test = np.array([55,60,65,70])
y_pred = model.predict(x_test)
```

```python
# Imprime los resultados
print('Tabla de multiplicar de 5:')
for i in range(len(x_test)):
    print('{} x 5 = {} (predicción: {})'.format(x_test[i], y_test[i], int(y_pred[i])))
```
