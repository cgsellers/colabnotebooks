GridSearchCV es una herramienta de la biblioteca Scikit-learn en Python que nos ayuda a ajustar los hiperparámetros de un modelo a través de una búsqueda exhaustiva en una "grilla" de posibles valores. A continuación, te muestro un ejemplo básico de cómo usar GridSearchCV con un modelo de regresión logística:

En este ejemplo, primero cargamos el conjunto de datos Iris y lo dividimos en un conjunto de entrenamiento y otro de prueba. A continuación, definimos una "parrilla" o "Grid" de posibles valores para los hiperparámetros C y penalty del modelo de regresión logística. Luego, usamos GridSearchCV para ajustar los hiperparámetros del modelo y encontrar la mejor combinación de hiperparámetros según la validación cruzada. Finalmente, usamos el modelo con los mejores hiperparámetros para hacer predicciones en el conjunto de prueba.

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

# cargamos el dataset Iris
iris = load_iris()

# dividimos los datos en un conjunto de entrenamiento y otro de prueba
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)

# definimos los posibles valores de los hiperparámetros que queremos ajustar
param_grid = {'C': [0.01, 0.1, 1, 10, 100], 'penalty': ['l1', 'l2']}

# definimos el modelo de regresión logística
model = LogisticRegression()

# usamos GridSearchCV para ajustar los hiperparámetros del modelo
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# imprimimos los mejores hiperparámetros encontrados
print("Mejor combinación de hiperparámetros: ", grid_search.best_params_)

# usamos el modelo con los mejores hiperparámetros para hacer predicciones en el conjunto de prueba
y_pred = grid_search.predict(X_test)
```
