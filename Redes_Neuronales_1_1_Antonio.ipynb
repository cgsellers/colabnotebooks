#2.1 Red Neuronal de puerta AND en Keras
---
---


Una red neuronal es un modelo computacional inspirado en la forma en que funciona el cerebro humano. Consiste en una serie de capas de nodos interconectados, donde cada nodo recibe una entrada, realiza una operación matemática y produce una salida que se conecta a los nodos de la siguiente capa.

Keras es una biblioteca de Python de código abierto para la creación de redes neuronales. Proporciona una interfaz de alto nivel para definir, entrenar y evaluar modelos de redes neuronales de forma fácil y rápida.

La función básica de una red neuronal en Keras se define mediante el objeto Sequential. Por ejemplo, una red neuronal con una capa oculta y una neurona  se puede definir así:

```python
#Importamos el objeto Sequential de Keras y la clase Dense de Keras
from keras.models import Sequential
from keras.layers import Dense
#Creamos el mdelo de red neuronal con una capa y una neurona, dos entradas y una salida
model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))
```

```python
# Definimos  x_train
x_train = [
                [0,0],
                [0,1],
                [1,0],
                [1,1]
]
x=x_train;
# Definimos y_train
y_train = [
                [0],
                [0],
                [0],
                [1]
]
```

```python
# como ves, las entradas de las listas estan ordenadas para que cuadren los 
# datos y las soluciones
print("Soluciones de una puerta logica AND:")
for i, d in enumerate(x_train):
    print("Para la entrada ", d, " la puerta da una salida de ", y_train[i])
```

---
---
Una función de pérdida es uno de los dos argumentos necesarios para compilar un modelo de Keras.

---
---

En el aprendizaje profundo, la pérdida se calcula para obtener los gradientes con respecto a los pesos del modelo y actualizar esos pesos en consecuencia a través de la retropropagación. Se calcula la pérdida y la red se actualiza después de cada iteración hasta que las actualizaciones del modelo no traigan ninguna mejora en la métrica de evaluación deseada.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])


model.fit(x_train, y_train, epochs=1000, batch_size=32)
```

```python
import numpy as np
# Transformaremos estas listas a listas de Numpy
x_train = np.array(x_train)
y_train = np.array(y_train)# 6)
# Por ultimo, ahora que nuestra red esta entrenada, vamos a probar a pasarle 
# datos a ver que es lo que nos saca. Para ello usaremos el modelo (la variable)
# como si fuera una funcion, pasandole por parametros una lista con los datos 
# que queremos que prediga

# Los datos sera nuestra lista de datos de entrada x_train
predicciones = model(x_train).numpy()

# Mostramos la salida de la red para cada uno de los datos
for i, p in enumerate(predicciones):
    print("Para la entrada ", x_train[i], " la red predice una salida de ", p)

# Banalizamos la salida de la red para cada uno de los datos
for i, p in enumerate(predicciones):
  if (p<0.5):
    print("Para la entrada ", x_train[i], " la red predice una salida de ", 0 ," i ",i)
else: print("Para la entrada ", x_train[i], " la red predice una salida de ", 1," i ",i)
```

```python
# Guardar predicciones para futuras verificaciones
predictions = model.predict(x_train)
```
