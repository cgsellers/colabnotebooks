```python
import pandas as pd
from fastavro import writer, parse_schema
```

```python
foo=pd.read_csv('2016.csv')
foo=pd.DataFrame(foo['ORIGIN'])
schema = {
    'name': 'Origen',
    'namespace' : 'CArlos',
    'type': 'record',
    'fields': [
        {'name': 'ORIGIN', 'type': 'string'}
    ]
}
```

```python
schemaParseado = parse_schema(schema)
```

```python
records = foo.to_dict('records')

# 3. Persistimos en un fichero avro
with open('origenes.avro', 'wb') as f:
    writer(f, schemaParseado, records)
```
