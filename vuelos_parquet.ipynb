```python
import pandas as pd
from fastavro import writer, parse_schema

# Leemos el csv mediante pandas
df = pd.read_csv('2009_cortado.csv')
# Limpiamos los datos (strip a los códigos postales) y nos quedamos con Alemania
df['DEST'] = df['DEST'].str.strip()
df['DEST']
```

```python
filtro = df['DEST']=="DCA"
df = df[filtro]

# 1. Definimos el esquema
schema = {
    'name': '2009_cortado',
    'namespace' : 'Carlos G',
    'type': 'record',
    'fields': [
        {'name': 'FL_DATE', 'type': 'string'},
        {'name': 'OP_CARRIER', 'type': 'string'},
        {'name': 'OP_CARRIER_FL_NUM', 'type': 'int'},
        {'name': 'ORIGIN', 'type': 'string'},
        {'name': 'DEST', 'type': 'string'},
        {'name': 'CRS_DEP_TIME', 'type': 'int'}
    ]
}
df
```

```python
schemaParseado = parse_schema(schema)

# 2. Convertimos el Dataframe a una lista de diccionarios
records = df.to_dict('records')

# 3. Persistimos en un fichero avro
with open('2009_cortado.avro', 'wb') as f:
    writer(f, schemaParseado, records)
```
