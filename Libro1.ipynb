```python
import pandas as pd
import numpy as np
```

```python
libro='Libro1.csv'
df=pd.read_csv(libro)
df
```

```python
df['nombre']=pd.Categorical(df['nombre'])
df['estado-civil']=pd.Categorical(df['estado-civil'])
df['sexo']=pd.Categorical(df['sexo'])
df['nacionalidadedad']=pd.Categorical(df['nacionalidadedad'])
df['enfermedad']=pd.Categorical(df['enfermedad'])
```

```python
df['nombre']=df['nombre'].cat.codes
df['estado-civil']=df['estado-civil'].cat.codes
df['sexo']=df['sexo'].cat.codes
df['nacionalidadedad']=df['nacionalidadedad'].cat.codes
df['enfermedad']=df['enfermedad'].cat.codes
```

```python
df.info()
```

```python
x=df.drop(['enfermedad'],axis=1)
y=df['enfermedad']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y, test_size=0.2,random_state=0)
```

#Regresión linear

```python
from sklearn import linear_model
```

```python
modelo=linear_model.LinearRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={"n_jobs":[-1,0,1,2]}
grid= GridSearchCV(estimator=modelo,param_grid=parameters,cv=3)
grid.fit(X_train,Y_train)
```

```python
X_test
```

```python
Y_test_pred=grid.predict(X_test)
```

```python
Y_test_pred
```

```python
import sklearn.metrics as metrics
```

```python
umbral=0.5
y_umbralizadas=1*(Y_test_pred[:]>umbral)
y_umbralizadas
```

```python
print(u"Matriz de confusión\n",metrics.confusion_matrix(Y_test, y_umbralizadas))
```
