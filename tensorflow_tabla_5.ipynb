```python
import tensorflow as tf
import numpy as np
```

```python
entrada=np.array([[5,0],[5,1],[5,2],[5,3],[5,4],[5,5],[5,6],[5,7],[5,8],[5,9],[5,10]], "int32")
input_data = tf.constant([[5,0],[5,1],[5,2],[5,3],[5,4],[5,5],[5,6],[5,7],[5,8],[5,9],[5,10]], dtype=tf.float32)
output_data = tf.constant([[0], [5], [10], [15],[20],[25],[30],[35],[40],[45],[50]], dtype=tf.float32)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1, input_shape=(2,)))
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(input_data, output_data, epochs=5000, verbose=True)
```

```python
x_test = np.array([[5,11],[5,12],[5,13],[5,14]], "int32")
y_test = np.array([55,60,65,70])
y_pred = model.predict(x_test)
```

```python
print('Tabla de multiplicar de 5:')
for i in range(len(x_test)):
    print('{} x 5 = {} (predicción: {})'.format(x_test[i], y_test[i], int(y_pred[i])))
```
