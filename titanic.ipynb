```python
import pandas as pd
```

```python
titanic='titanic.csv'
```

```python
#Genera un dataframe y muéstralo.
df=pd.read_csv(titanic)
df
```

```python
#Mostrar por pantalla las filas pares del DataFrame.
filas_pares=df[::2]
filas_pares
```

```python
#Mostrar por pantalla los datos del pasajero con identificador 666
id_666=df[df["PassengerId"]==666]
id_666
```

```python
#Muestra aquellas filas de los pasajeros que tengan el campo sexo a male o female y muestra los resultados del fichero tras esta operación.
hombres_mujeres=df[(df["Sex"]=='male')|(df["Sex"]=='female')]
hombres_mujeres
```

```python
#A todos los pasajeros que tengan su nombre a NULL, hay que asignarles el valor de “anonymus”.
df['Name'].replace({"NULL","anonymus"})
```

```python
#Muestra los menores de 53 años: Los primeros 10 y el número total. También los datos de esta extracción.
Mayores_53=df[df['Age']<53]
Mayores_53.head(n=10)
```

```python
Mayores_53.count()
```

```python
#Muestra todos los datos de los viajeros de clase 2 y 3, y de éstos los menores de 25 años. Los 6 primeros y el número total, así como los datos de esta extracción.
clases_2_3=df[(df["Pclass"] == 2) | (df["Pclass"] == 3)]
menores25=clases_2_3[clases_2_3['Age']<25]
menores25.head(n=6)
```

```python
menores25.count()
```
