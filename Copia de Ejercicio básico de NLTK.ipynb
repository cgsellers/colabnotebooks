1. Importa la librería NLTK y descarga el corpus 'gutenberg'.

```python
#Write your code here
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import spacy
nltk.download('gutenberg')
```

2. Carga el texto del libro "Sense and Sensibility" de Jane Austen.

```python
#Write your code here
texto=nltk.corpus.gutenberg.raw('austen-sense.txt')
type(texto)
```

3. Tokeniza el texto. Muestra los tokens.

```python
#Write your code here

nltk.download('punkt')
tokens = word_tokenize(texto)
tokens=[word.lower() for word in tokens if word.isalpha()]
print(tokens)
```

4. Calcula la frecuencia de cada palabra en el texto. También una pequeña gráfica de las mismas.

```python
#Write your code here
freq = nltk.FreqDist(tokens)
for key,val in freq.items():
    print (str(key) + ': ' + str(val))
```

```python
#Write your code here
import seaborn as sns
sns.set()
freq.plot(50,cumulative=False)
```

5. Muestra las 10 palabras más frecuentes en el texto. También una pequeña gráfica de las mismas.

```python
list(freq)[:10]
```

```python
freq.plot(10,cumulative=False)
```

6. Elimina las stopwords.

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

clean_tokens = tokens[:]
 
for token in tokens:
 
    if token in stopwords.words('english'):
 
        clean_tokens.remove(token)
```

6.1. ¿Cómo añadirías las palabras vacías al conjunto estándar?


```python
tokens.extend(stopwords.words('english'))
```

6.2. ¿Cómo eliminarías las palabras vacías del conjunto estándar?.

```python
#Write your code here
tokens=[x for x in tokens if x not in stopwords.words('english')]
```

6.3. ¿Cómo crearías un conjunto propio de palabras vacías?.

```python
#Write your code here
vacias=word_tokenize(str(stopwords.words('english')))
vacias
```

7. Aplica stemming. Muestra los tokens resultantes.

```python
#Write your code here
stemmer = PorterStemmer()
clean_tokens_stems = [stemmer.stem(token) for token in clean_tokens]
clean_tokens_stems
```

8. Aplica lematización. Muestra los tokens resultantes.

```python
#Write your code here
lem = spacy.blank('en')(' '.join(clean_tokens_stems))
clean_tokens_lem=[x.text for x in lem]
clean_tokens_lem
```

9. Indica al menos tres consideraciones que podemos sacar el texto analizado.

```python
freq = nltk.FreqDist(clean_tokens_lem)
for key,val in freq.items():
    print (str(key) + ':' + str(val))
```

Viendo la frecuencia de las raíces, las más comunes son sustantivos propios (dashwood, elinor, mariann) y comunes(sister, mother, man, one), seguidas por raíces que se pueden utilizar tanto en verbos como sustantivos (look)
