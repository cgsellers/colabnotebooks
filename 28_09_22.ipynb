```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
plt.plot([i**2 for i in x]) #va a mostrar los cuadados del 1 al 9 (1 a 81)
plt.axis([1,10,0,81])
plt.show()
```

```python
x = np.arange(2,40,3)
plt.plot(x)
plt.show()
plt.hist(x)   #histograma de x
```

```python
x = np.random.rand(1000)
plt.hist(x,20, color="cadetblue")
plt.xlabel('Edades')
plt.ylabel('Salario')
plt.title('Sueldos')
plt.show()
```

```python
x = range(10)
y = []
for i in x:
 i=i*2
 y.append(i)

plt.scatter(x,y, color='cadetblue') #con scatter nos muestra los valores de manera individual
plt.xlabel('X')
plt.ylabel('Y')
plt.title('distribución')
plt.show()
```
